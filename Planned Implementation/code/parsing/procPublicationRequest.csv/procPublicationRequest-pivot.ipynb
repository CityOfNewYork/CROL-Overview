**Message Count in descending order, grouped by Agency name & TypeOfNoticeDescription**

```python
import pandas as pd
pd.options.display.max_rows=2
```

```python
fn = 'procPublicationRequest Oct-Dec 2014 (Updated) - Sheet1-2.csv'
df = pd.read_csv(fn, header=0)
```

```python
table = df[["AgencyName", "TypeOfNoticeDescription", "RequestID"]]
table = pd.DataFrame(table.groupby(["AgencyName", "TypeOfNoticeDescription"])["RequestID"].count())
table
```

```python
# Rename Column 
table.columns = ['Message Count']
table = table.sort(['Message Count'], ascending=[0])
table
```

```python
# Expand display and show it
pd.options.display.max_rows=150
table
```
