**Record count sorted and indexed by Agency name & TypeOfNoticeDescription**

```python
import pandas as pd
pd.options.display.max_rows=10
```

```python
fn = 'procPublicationRequest Oct-Dec 2014 (Updated) - Sheet1-2.csv'
df = pd.read_csv(fn, header=0)
```

```python
df["TypeOfNoticeDescription"] = df["TypeOfNoticeDescription"].astype("category")
df["TypeOfNoticeDescription"].cat.set_categories(["Public Hearings","Notice","Meeting","Sale"],inplace=True)
```

```python
table = pd.pivot_table(df, index=["AgencyName", "TypeOfNoticeDescription"], 
    values=['RequestID'], aggfunc=[len], 
    fill_value=0, margins=True)
```

```python
table
```

```python
# stack and remove non zero length counts
stacked = table.stack()
stacked = stacked[stacked.len>0]
stacked
```

```python
# stacked = stacked.unstack()
sorted = stacked.sort(['len'], ascending=[0])
```
